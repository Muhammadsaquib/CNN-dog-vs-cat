```python
import cv2
import tensorflow as tf
import matplotlib.pyplot as plt


CATA = ["Cat","Dog"]

def prepare(filename):
    img_size = 60
    img_array = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)
    plt.imshow(img_array, cmap = "gray")
    plt.show()
    new_array = cv2.resize(img_array,(img_size, img_size))
    return new_array.reshape(-1, img_size, img_size, 1)
                           
model = tf.keras.models.load_model("Cats-vs-dogs-64x2-CNN time")
                           
                           
prediction = model.predict([prepare(r"C:\Users\tahas\anaconda3\dogg.jpg")])
print(CATA[int(prediction[0][0])])
```
